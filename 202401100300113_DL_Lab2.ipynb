```python
import pandas as pd
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x*(1-x)
```

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
```

```python
y = np.array([[0],[1],[1],[0]])
```

```python
input_size = 2
hidden_size = 4
output_size = 1
W1 = np.random.randn(input_size,hidden_size)
b1 = np.zeros((1,hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))
```

```python
learning_rate = 0.1
epochs = 5000
losses = []
```

```python
for epoch in range(epochs):
    z1 = np.dot(x, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    y_hat = sigmoid(z2)
    loss = np.mean((y - y_hat) ** 2)
    losses.append(loss)
    error_output = y_hat - y
    d_output = error_output * sigmoid_derivative(y_hat)
    error_hidden = np.dot(d_output, W2.T)
    d_hidden = error_hidden * sigmoid_derivative(a1)
    W2 -= learning_rate * np.dot(a1.T, d_output)
    b2 -= learning_rate * np.sum(d_output, axis=0, keepdims=True)
    W1 -= learning_rate * np.dot(x.T, d_hidden)
    b1 -= learning_rate * np.sum(d_hidden, axis=0, keepdims=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(losses)
plt.title("Training Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```
